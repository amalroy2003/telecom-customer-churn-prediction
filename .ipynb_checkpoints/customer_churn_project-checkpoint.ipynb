# CHECKING FOR OUTLIERS

```python
# Numerical features and churn relationship visualization
plt.figure(figsize=(12, 8))
# Monthly Charges 
plt.subplot(2, 3, 1)
sns.boxplot(x='Churn', y='MonthlyCharges', data=cus)
plt.title("Monthly Charges vs Churn")

# Tenure 
plt.subplot(2, 3, 2)
sns.boxplot(x='Churn', y='tenure', data=cus)
plt.title("Tenure vs Churn")

# Total Charges 
plt.subplot(2, 3, 3)
sns.boxplot(x='Churn', y='TotalCharges', data=cus)
plt.title("Total Charges vs Churn")

plt.tight_layout()
plt.show()
```

```python
# Visualize distributions before and after handling outliers
plt.figure(figsize=(15, 5))

for i, col in enumerate(['tenure', 'MonthlyCharges', 'TotalCharges'], 1):
    plt.subplot(1, 3, i)
    sns.histplot(cus[col], kde=True)
    plt.title(f"Distribution of {col} (Before Handling Outliers)")

plt.tight_layout()
plt.show()

# Calculating skewness
tenure_skewness = skew(cus['tenure'])
monthly_charges_skewness = skew(cus['MonthlyCharges'])
total_charges_skewness = skew(cus['TotalCharges'])

# Printing skewness values
print(f"Skewness of tenure: {tenure_skewness:.4f}")
print(f"Skewness of MonthlyCharges: {monthly_charges_skewness:.4f}")
print(f"Skewness of TotalCharges: {total_charges_skewness:.4f}")
```

```python
#CHECKING COUNT OF OUTLIERS
for col in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    for churn_value in cus["Churn"].unique():
        subset = cus[cus["Churn"] == churn_value]
        Q1 = subset[col].quantile(0.25)
        Q3 = subset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = subset[(subset[col] < lower_bound) | (subset[col] > upper_bound)]
        print(f"Outliers in {col} for Churn={churn_value}: {len(outliers)}")
```

### tenure

```python
# Define the IQR bounds
Q1 = cus.loc[cus["Churn"] == 1, "tenure"].quantile(0.25)
Q3 = cus.loc[cus["Churn"] == 1, "tenure"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Apply winsorization only to tenure outliers for Churn=1
cus.loc[(cus["Churn"] == 1) & (cus["tenure"] > upper_bound), "tenure"] = upper_bound
```

```python
# Tenure 
plt.figure(figsize=(3,4))
# plt.subplot(2, 3, 2)
sns.boxplot(x='Churn', y='tenure', data=cus)
plt.title("Tenure vs Churn")
```

```python
for col in ['tenure']:
    for churn_value in cus["Churn"].unique():
        subset = cus[cus["Churn"] == churn_value]
        Q1 = subset[col].quantile(0.25)
        Q3 = subset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = subset[(subset[col] < lower_bound) | (subset[col] > upper_bound)]
        print(f"Outliers in {col} for Churn={churn_value}: {len(outliers)}")
```

### TotalCharges

```python
# Apply log transformation first
cus['TotalCharges'] = np.log1p(cus['TotalCharges'])

# Recalculate bounds after log transformation
lower_bound = np.percentile(cus.loc[cus['Churn'] == 1, 'TotalCharges'], 1)
upper_bound = np.percentile(cus.loc[cus['Churn'] == 1, 'TotalCharges'], 99)

# Apply capping after log transformation
cus.loc[(cus['Churn'] == 1) & (cus['TotalCharges'] < lower_bound), 'TotalCharges'] = lower_bound
cus.loc[(cus['Churn'] == 1) & (cus['TotalCharges'] > upper_bound), 'TotalCharges'] = upper_bound
```

```python
pt = PowerTransformer(method='yeo-johnson')  # Box-Cox requires positive values, Yeo-Johnson doesn't
cus['TotalCharges'] = pt.fit_transform(cus[['TotalCharges']])
```

```python
for col in ['TotalCharges']:
    for churn_value in cus["Churn"].unique():
        subset = cus[cus["Churn"] == churn_value]
        Q1 = subset[col].quantile(0.25)
        Q3 = subset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = subset[(subset[col] < lower_bound) | (subset[col] > upper_bound)]
        print(f"Outliers in {col} for Churn={churn_value}: {len(outliers)}")
```

### MODEL TRAINING

#### Logistic without L2

```python
# Logistic Regression without Regularization
lr_no_reg = LogisticRegression()  # No regularization
lr_no_reg.fit(X_train, y_train)

y_pred_no_reg = lr_no_reg.predict(X_test)
print("Logistic Regression Accuracy (No Regularization):", accuracy_score(y_test, y_pred_no_reg))
print("\nClassification Report:\n", classification_report(y_test, y_pred_no_reg))
```

#### L2

```python
lr = LogisticRegression(penalty='l2', solver='liblinear')  # L2 Regularization
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
# print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr))
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Define the parameter grid
param_grid = {
    'max_depth': [3, 5, 10], 
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy'],  # Try different split criteria
    'min_samples_leaf': [1, 2, 4]  # Minimum samples in a leaf node
}

# Initialize Decision Tree model
dt = DecisionTreeClassifier(random_state=42)

# Perform Grid Search
dt_grid = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
dt_grid.fit(X_train, y_train)

# Get best model
best_dt = dt_grid.best_estimator_

# Predictions
y_pred_dt = best_dt.predict(X_test)

# Print best parameters
print("Best Decision Tree Parameters:", dt_grid.best_params_)

# Evaluate model
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))
```

#### Random Forest

```python
# Initialize default Random Forest model
rf_baseline = RandomForestClassifier(random_state=42)
rf_baseline.fit(X_train, y_train)

# Predictions
y_pred_rf_baseline = rf_baseline.predict(X_test)

# Evaluate performance
print("Random Forest (Baseline) Accuracy:", accuracy_score(y_test, y_pred_rf_baseline))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf_baseline))
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform grid search
rf_tuned = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
rf_tuned.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", rf_tuned.best_params_)

# Predictions using the best model
y_pred_rf_tuned = rf_tuned.best_estimator_.predict(X_test)

# Evaluate performance
print("\nRandom Forest (Tuned) Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf_tuned))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define hyperparameter grid
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Perform randomized search
rf_tuned = RandomizedSearchCV(RandomForestClassifier(random_state=42),
                              param_distributions=param_dist,
                              n_iter=20,  # Reduce for faster search, increase for better tuning
                              cv=5,
                              scoring='accuracy',
                              n_jobs=-1,
                              random_state=42)

rf_tuned.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", rf_tuned.best_params_)

# Predictions using the best model
y_pred_rf_tuned = rf_tuned.best_estimator_.predict(X_test)

# Evaluate performance
print("\nRandom Forest (Tuned) Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf_tuned))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Step 1: Define the parameter grid
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.1, 1.0, 10, 100],
    'solver': ['lbfgs', 'liblinear', 'saga'],
    'max_iter': [100, 200, 300],
    'l1_ratio': [None, 0.5]  # Only used if penalty is 'elasticnet'
}

# Step 2: Initialize the GridSearchCV with the model and parameter grid
grid_search = GridSearchCV(LogisticRegression(random_state=100), param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Step 3: Fit the GridSearchCV to the training data
grid_search.fit(xr_train, yr_train)

# Step 4: Get the best hyperparameters and evaluate the model
best_model = grid_search.best_estimator_
y_pred_smote = best_model.predict(xr_test)

print(f"Best Hyperparameters: {grid_search.best_params_}")
print(classification_report(yr_test, y_pred_smote, labels=[0, 1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Step 1: Define the parameter grid for RandomForestClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Step 2: Initialize the GridSearchCV with the RandomForestClassifier and parameter grid
grid_search = GridSearchCV(RandomForestClassifier(random_state=100), param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Step 3: Fit the GridSearchCV to the training data
grid_search.fit(xr_train, yr_train)

# Step 4: Get the best hyperparameters and evaluate the model
best_model = grid_search.best_estimator_
y_pred_smote = best_model.predict(xr_test)

print(f"Best Hyperparameters: {grid_search.best_params_}")
print(classification_report(yr_test, y_pred_smote, labels=[0, 1]))
```
